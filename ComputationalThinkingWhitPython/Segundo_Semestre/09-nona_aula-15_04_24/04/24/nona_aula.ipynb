```python
# Dataset sobre falha de maquinas

dados = pd.read_csv(r"./manutencao_preditiva.csv")
dados
```

```python
"""
Exercícios/Exemplos:

1 - crie um programa para somar as variáveis: temperatura do ar e processos e retorne uma nova variavel

2 - Faça a soma das temperaduras do ar e do processo apenas para as maquinas tipo M

3 - Mostre qual a média, mediana, min e max das temperaturas quando huver apenas máquinas do tipo H.



"""
```

```python
# 1

dados['Soma Temperaturas'] = dados.apply(lambda row: row['Temperatura Ar [K]'] + row['Temperatura Processo [K]'], axis=1)
dados
```

```python
# 2

dados_tipo_m = dados[dados['Tipo']=='M']
dados_tipo_m['Soma'] = dados_tipo_m.apply(lambda row: row['Temperatura Ar [K]'] + row['Temperatura Processo [K]'], axis=1)
dados_tipo_m
```

```python
# 3

dados_tipo_h = dados[dados['Tipo']=='H']

print('Média da Temperatura Ar: ', dados_tipo_h['Temperatura Ar [K]'].apply(lambda x: x).mean())
print('Média da Temperatura de Processos: ', dados_tipo_h['Temperatura Processo [K]'].apply(lambda x: x).mean())

print('Mediana da Temperatura Ar: ', dados_tipo_h['Temperatura Ar [K]'].apply(lambda x: x).median())
print('Mediana da Temperatura de Processos: ', dados_tipo_h['Temperatura Processo [K]'].apply(lambda x: x).median())

print('Mínima da Temperatura Ar: ', dados_tipo_h['Temperatura Ar [K]'].apply(lambda x: x).min())
print('Mínima da Temperatura de Processos: ', dados_tipo_h['Temperatura Processo [K]'].apply(lambda x: x).min())

print('Máxima da Temperatura Ar: ', dados_tipo_h['Temperatura Ar [K]'].apply(lambda x: x).max())
print('Máxima da Temperatura de Processos: ', dados_tipo_h['Temperatura Processo [K]'].apply(lambda x: x).max())

```

```python
# Normalização de Dados

df_sample = pd.DataFrame({
    'A':[6,5,8,11],
    'B':[100,122,156,189],
    'indice': [1,2,3,4]
})

x = df_sample['indice']
y1 = df_sample['A']
y2 = df_sample['B']

plt.plot(x, y1, x, y2)
plt.show()
```

```python
df_normalizado = df_sample.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df_normalizado
```

```python
# Faça a normalização (0 e 1) apenas para varíavel Velocidade Rotacao [rpm]

dados_vel_rotacao = dados[['Velocidade Rotacao [rpm]']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
dados_vel_rotacao
```

```python
# Aplicar apply para verificar comprimento de dados (strings)

df = pd.DataFrame({
    'Texto':['Ola', 'mundo', 'python', 'pandas']
})

df['Contagem'] = df['Texto'].apply(lambda texto: len(texto))
df
```

```python
# Criar uma condição para inserir uma coluna chamada comprimento_texto e se o comprimento for mair do qe 5, insir LONGO, caso contrario CURTO
df['Compirmento Texto'] = df['Texto'].apply(lambda texto: 'Longo' if len(texto) > 5 else 'Curto')

df
```

```python
# Crie um programa para verificar o comprimento do texto na coluna Tipo de Falha. Após isso, insira uma nova coluna mosttrando. Se o comprimento menor ou igual a 10- curto. Se o comprimento maior que 10 e meno que 15 - médio. Se o compirmento maior que 15 - longo.

dados['Comprimento texto'] = dados['Tipo da Falha'].apply(lambda texto: 'Curto' if len(texto) <= 10 else 'Médio' if 10 < len(texto) <= 15 else 'Longo')

dados
```

```python
# Utilizando groupby do pandas

df_new = pd.DataFrame({
    'var_1':['A','A','B','A','C','C','A','B','A','B'],
    'var_2':[100,122,156,189,50,63,9,15,78,10],
    'var_3':[10,22,16,89,5,3,9,15,8,10]
})

df_new
```

```python
# reset_index() retorna ao formato dataframe

agrupado_media = df_new.groupby('var_1').mean().reset_index()
agrupado_media
```

```python
agrupado_mediana = df_new.groupby('var_1').median().reset_index()
agrupado_mediana
```

```python
agrupado_minima = df_new.groupby('var_1').min().reset_index()
agrupado_minima
```

```python
agrupado_maxima = df_new.groupby('var_1').max().reset_index()
agrupado_maxima
```
