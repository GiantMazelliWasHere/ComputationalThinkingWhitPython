## 8. Crie uma estrutura para mostrar as informaÃ§Ãµes de velocidade de rotaÃ§Ã£o e torqe. PorÃ©m, escreva 'tempratura alta de processo' quando a temperatura do processo for maior que  mÃ©dia e a 'temperatura baixa de processo' quando for menor que a mÃ©dia

```python
media = df['Temperatura Processo [K]'].mean()

for index, row in df.iterrows():
    if row['Temperatura Processo [K]'] > media:
        print(f'Linha {index}:')
        print(f"Velocidade de rotaÃ§Ã£o: {row['Velocidade Rotacao [rpm]']} e Torque: {row['Torque [Nm]']}")
        print('Temperatura alta de processo')
    else:
        print(f'Linha {index}:')
        print(f"Velocidade de rotaÃ§Ã£o: {row['Velocidade Rotacao [rpm]']} e Torque: {row['Torque [Nm]']}")
        print('Temperatura baixa de processo')
```

## 9. Crie um programa para identificar e mostrar a quantidade de falhas no conjnto de dados.

```python
contagem_falhas = {}

for index, row in df.iterrows():
    tipo_falha = row['Tipo da Falha']

    if tipo_falha in contagem_falhas:
        contagem_falhas[tipo_falha] += 1

    else:
        contagem_falhas[tipo_falha] = 1

for tipo_falha, contagem in contagem_falhas.items():
    print(f"Tipo da falha: {tipo_falha} - Quantidade: {contagem}")
    
```

## 11. Crie um programa para apresentar as informaÃ§Ãµes de UDI e ID produto quando houver apenas falhas do tipo no failure.

## 12. Crie um programa para apresentar a quantidade de falhas no conjunto de dados de falhas em mÃ¡quinas usando estrtura while

```python
contagem_falhas = {}
iterator = df.iterrows()

while True:
    try:
        index, row = next(iterator)
        tipo_falha = row['Tipo da Falha']

        if tipo_falha in contagem_falhas:
            contagem_falhas[tipo_falha] += 1
        else:
            contagem_falhas[tipo_falha] = 1

    except StopIteration:
        break


for tipo_falha, contagem in contagem_falhas.items():
    print(f"Tipo da falha: {tipo_falha} - Quantidade: {contagem}")
```

## Calculando a mÃ©dia usando for e while

```python
#for

soma_desgaste = 0
count_m = 0

for index, row in df.iterrows():
    if row['Tipo'] == 'M':
        soma_desgaste += row['Desgaste Ferramenta [min]']
        count_m += 1
    
if count_m > 0:
    media_desgaste = soma_desgaste / count_m
    print(f"A mÃ©dia de desgaste da ferramente para o tipo M Ã©: {media_desgaste}")
else:
    print('NÃ£o hÃ¡ dados')
```

```python
#while

soma_desgaste = 0
count_m = 0
iterator = df.iterrows()

while True:
    try:
        index, row = next(iterator)
        if row['Tipo'] == 'M':
            soma_desgaste += row['Desgaste Ferramenta [min]']
            count_m += 1
    except StopIteration:
        break
    
if count_m > 0:
    media_desgaste = soma_desgaste / count_m
    print(f"A mÃ©dia de desgaste da ferramente para o tipo M Ã©: {media_desgaste}")
else:
    print('NÃ£o hÃ¡ dados')
```
