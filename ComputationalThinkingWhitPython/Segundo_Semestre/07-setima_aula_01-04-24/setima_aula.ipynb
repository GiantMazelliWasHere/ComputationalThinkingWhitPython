```python
# Estrutura de repetição - ex06
# utilizar o while para calcular estatísticas descritivas de colunas numéricas, mas iterrompendo o loop quando um determinado critério é atingido

dados={'Nome' : ['Alice', 'Bob','Charlie','David','Emily'],
       'Idade'  : [25,30,35,40,45],
       'Salário' : [5000,6000,7000,8000,9000]}

df = pd.DataFrame(dados)
df
```

```python
# Média Movel

soma_salarios = 0
contador = 0
i = 0

while i < len(df):
    soma_salarios += df['Salário'][i]
    contador += 1
    media_salarios = soma_salarios / contador
    print(f'A média salário é {media_salarios}')
    if media_salarios > 7000:
        break
    i+=1
```

```python
# Estrutura de repetição - ex07
# Calcular bonus para o funcionario usando função def

# crie um programa para calcular bonus de 10% para os funcionarios, mas mostre o bônus para cada funcionario

def calcula_bonus(salario):
    return salario * 0.1

bonus = []

for indice, linha in df.iterrows():
    bonus.append(calcula_bonus(linha['Salário']))

print('Bônus', bonus)
```

```python
# Estrutura de repetição - ex08

# calcule a média de idade e salário e retorne apenas um dataframe contendo as duas informações

media_idade = df['Idade'].mean()
media_salario = df['Salário'].mean()

df_final = {'Média Idade': [media_idade],
            'Média Salário' : [media_salario]}

df_final = pd.DataFrame(df_final)
df_final
```

1. Utilizando o dataframe sobre falhas em máquinas, crie um programa para retornar a média das colunas com valores númericos e não numéricos fazer a contagem das informações que se repetem.

```python
def media(df):
    for coluna in df.select_dtypes(include = ['int', 'float']):
        media = df[coluna].mean()
        print(f'A média de {coluna}: {media}')

def soma_valores(df):
    for coluna in df.select_dtypes(include=['object']):
        contagem_valores = df[coluna].value_counts()
        print(f'Contagem de valores únicoes em {coluna}: {contagem_valores}')

def operacoes(df):
    media(df)
    soma_valores(df)

operacoes(dados)
```
