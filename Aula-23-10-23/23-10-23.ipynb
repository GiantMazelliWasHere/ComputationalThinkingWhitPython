```python
data = {'Nome': ['Alice', 'Bob', 'Charlie', 'Ana', 'Alice', 'Bob'], 
        'Idade': [25, 30, 22, 22, 38, 35], 
        'País': ['Brasil', 'Inglaterra', 'Estados Unidos', 'Itália', 'Austrália', 'México'],
}
covid = []
indice = 0

df = pd.DataFrame(data)

while indice < len(df):
    pais = df['País'][indice]
    if pais == 'Brasil' or pais == 'Itália':
        covid.append("Tem Covid.")
    else:
        covid.append("Não tem Covid.")
    indice += 1

df['Covidou?'] = covid

df
```

```python
vacinas = []
prioridade = []

indice = 0

while indice < len(df):
    vacina = df['Covidou?'][indice]
    if vacina == 'Tem Covid.':
        vacinas.append("Descanca desgramado")
    else:
        vacinas.append("Hora de vacinar. Att: Zé Gotinha")
    
    idade = df['Idade'][indice]
    if idade > 35:
        prioridade.append('Prioridade alta')
    else:
        prioridade.append('Prioridade baixa')

    indice +=1

df['Diagnóstico'] = vacinas
df['Prioridade'] = prioridade

df
```
